```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import astropy.io.ascii
```

```python
maukea = astropy.io.ascii.read("mktrans_zm_10_10.dat.txt",format = "no_header")
f09 = astropy.io.ascii.read("f098m.IR.tab.txt",format = "no_header")
f10 = astropy.io.ascii.read("f105w.IR.tab.txt",format = "no_header")
f11 = astropy.io.ascii.read("f110w.IR.tab.txt",format = "no_header")
f12 = astropy.io.ascii.read("f125w.IR.tab.txt",format = "no_header")
f14 = astropy.io.ascii.read("f140w.IR.tab.txt",format = "no_header")
f16 = astropy.io.ascii.read("f160w.IR.tab.txt",format = "no_header")
```

```python
mauwave = []
mautrans = []
f9wave = []
f9trans = []
f10wave = []
f10trans = []
f11wave = []
f11trans = []
f12wave = []
f12trans = []
f14wave = []
f14trans = []
f16wave = []
f16trans = []
for x in maukea:
    if x[0] >= 0.9 and x[0] <= 2.7:
        mauwave.append(x[0])
        mautrans.append(x[1])
for b in f09:
    if b[1] >= 9000 and b[1] <= 27000:
        f9wave.append(b[1]/10000)
        f9trans.append(b[2])
for a in f10:
    if a[1] >= 9000 and a[1] <= 27000:
        f10wave.append(a[1]/10000)
        f10trans.append(a[2])
for c in f11:
    if c[1] >= 9000 and c[1] <= 27000:
        f11wave.append(c[1]/10000)
        f11trans.append(c[2])
for d in f12:
    if d[1] >= 9000 and d[1] <= 27000:
        f12wave.append(d[1]/10000)
        f12trans.append(d[2])
for e in f14:
    if e[1] >= 9000 and e[1] <= 27000:
        f14wave.append(e[1]/10000)
        f14trans.append(e[2])
for f in f16:
    if f[1] >= 9000 and f[1] <= 27000:
        f16wave.append(f[1]/10000)
        f16trans.append(f[2])
```

```python
fig = plt.figure(num=None,figsize=(6,5.7))
ax = fig.add_axes([0.9,0,2.75,1])
plt.plot(mauwave,mautrans)
plt.plot(f9wave,f9trans,label = "9")
plt.plot(f10wave,f10trans,label = '10')
plt.plot(f11wave,f11trans,label = '11')
plt.plot(f12wave,f12trans,label = '12')
plt.plot(f14wave,f14trans,label = '14')
plt.plot(f16wave,f16trans,label = '16')
plt.xlim(0.9,2.75)
plt.title("Atmospheric Transmission Curve in Near-IR vs. HST IR-Filters")
plt.xlabel("Wavelength [$\mu$m]")
plt.ylabel("Transmission")
#plt.annotate(s='this',xy=(1,.5),xytext=(1,1))
#So the arrows aren't in order, but it'll work
arrow1 = patches.FancyArrowPatch((0.9,0.05),(1.079,0.05),arrowstyle='<->',color = 'orange',zorder = 1000,mutation_scale=20)
arrow2 = patches.FancyArrowPatch((1.09,0.05),(1.41,0.05),arrowstyle='<->',color = 'purple',zorder = 1000,mutation_scale=20)
arrow3 = patches.FancyArrowPatch((0.9,0.25),(1.202,0.25),arrowstyle='<->',color = 'green',zorder = 1000,mutation_scale=20)
arrow4 = patches.FancyArrowPatch((0.9,0.15),(1.403,0.15),arrowstyle='<->',color = 'red',zorder = 1000,mutation_scale=20)
arrow5 = patches.FancyArrowPatch((1.202,0.25),(1.6,0.25),arrowstyle='<->',color = 'brown',zorder = 1000,mutation_scale=20)
arrow6 = patches.FancyArrowPatch((1.39,0.15),(1.692,0.15),arrowstyle='<->',color = 'violet',zorder = 1000,mutation_scale=20)
ax.add_patch(arrow1)
ax.add_patch(arrow2)
ax.add_patch(arrow3)
ax.add_patch(arrow4)
ax.add_patch(arrow5)
ax.add_patch(arrow6)
ax.annotate(s='F098M',xy=(0.97,0.06),color='orange')
ax.annotate(s='F125W',xy=(1.23,0.06),color='purple')
ax.annotate(s='F105W',xy=(1.01,0.26),color='green')
ax.annotate(s='F110W',xy=(1.01,0.16),color='red')
ax.annotate(s='F140W',xy=(1.25,0.26),color='brown')
ax.annotate(s='F160W',xy=(1.52,0.16),color='violet')
ax.annotate(s='F098M',xy=(0.97,0.06),color='orange')
ax.annotate(s='Atmospheric Transmission',xy=(2.50,1),color='b')
```
